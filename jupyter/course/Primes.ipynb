```python
import sympy
from collections import defaultdict
import pandas as pd
```

```python
LIMIT = 10**3
i = 1
primes = defaultdict(int)
lastprime = 0
while lastprime < LIMIT:
    lastprime = sympy.ntheory.prime(i)
    primes[lastprime] = 1
    i = i + 1
data = {i: primes[i] for i in range(LIMIT)}

s1 = pd.Series(data.values())
s1.plot(figsize=(20, 10))
```
